```python
from matplotlib import pylab as plt
import numpy as np
from scipy.stats import norm
#%matplotlib notebook
%matplotlib inline
```

```python
rho_m = np.load('mc_data/crw_rho_m.npy')

xstart = np.mean(rho_m)-5*np.std(rho_m)
xstop = np.mean(rho_m)+5*np.std(rho_m)
X_plot = np.linspace(xstart, xstop, 1000)#[:, np.newaxis]
log_dens = norm.pdf(X_plot,np.mean(rho_m),np.std(rho_m))
plt.plot(X_plot, (log_dens),'k-',linewidth=2)
plt.hist(rho_m,normed=True,range=[xstart,xstop],bins=50,alpha=0.4)
plt.fill_between(X_plot,np.zeros_like(X_plot), log_dens,alpha=0.4)#, fc='#AAAAFF')

plt.xlim(xstart,xstop)
plt.show()
```

```python
rho_s=np.load('mc_data/cm_rho_s.npy')
alpha=np.load('mc_data/cm_alpha.npy')

ia = np.load('mc_data/cm_ia.npy')
il = np.load('mc_data/cm_il.npy')
ig = np.load('mc_data/cm_ig.npy')

allData = [rho_s,alpha,ia,il,ig]

for i in range(len(allData)):
    
    pltThis = allData[i]
    xstart = np.mean(pltThis)-5*np.std(pltThis)
    xstop = np.mean(pltThis)+5*np.std(pltThis)
    X_plot = np.linspace(xstart, xstop, 1000)#[:, np.newaxis]
    log_dens = norm.pdf(X_plot,np.mean(pltThis),np.std(pltThis))
    plt.plot(X_plot, (log_dens),'k-',linewidth=2)
    plt.hist(pltThis,normed=True,range=[xstart,xstop],bins=50,alpha=0.4)
    plt.fill_between(X_plot,np.zeros_like(X_plot), log_dens,alpha=0.4)#, fc='#AAAAFF')

    plt.xlim(xstart,xstop)
    plt.show()
    plt.figure()

plt.plot(ia)
print(np.median(rho_s))
print(np.median(alpha))
print(np.median(ia))
print(np.median(il))
print(np.median(ig))
```
